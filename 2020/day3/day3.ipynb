```python
with open('input.txt' ,'r') as fh:
    data = fh.read().splitlines()
```

```python
data[:4]
```

```python
slope = list(zip(data, data[1:] + data[:1])) 
```

```python
slope[:6]
```

```python
h = 0
num_trees = 0

for i, j in slope:
    h += 3
    if h >= len(j):
        h -= len(j)
    hit = j[h]
    if hit == '#':
        num_trees += 1
        
print(num_trees)
```
